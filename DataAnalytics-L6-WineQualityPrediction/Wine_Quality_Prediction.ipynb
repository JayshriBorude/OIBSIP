# Class Imbalance Discussion

The distribution of wine quality scores is examined to determine whether certain quality classes are underrepresented.

An imbalanced dataset can bias the model toward majority classes, reducing its ability to correctly classify minority quality levels. To simplify the classification task and improve model performance, quality scores will be grouped into broader categories during feature engineering.

```python
df["quality"] = df["quality"].apply(
    lambda x: 1 if x >= 7 else 0
)

df["quality"].value_counts()
```

# Feature Engineering

The original quality scores range from 3 to 8. To simplify the classification task and reduce class imbalance, quality scores are converted into binary classes:

- Good Wine (1): Quality ≥ 7
- Bad Wine (0): Quality < 7

```python
df["quality"] = df["quality"].apply(lambda x: 1 if x >= 7 else 0)

print(df["quality"].value_counts())
```
